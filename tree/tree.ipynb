```python
#blue print of a tree
class BST():
    def __init__(self, data):
        self.key = data
        self.left = None
        self.right = None
#=====================================================================================================================================
# insertion operation
# condition1: if tree is empty then given data will be root node
    def insert(self,data):
        if not self.key:
            self.key = data
            return
# condition2: ignoring duplicate value
        if self.key == data:
            return
# condition3: 1. if given node is less than the root node then first check if left node is present:
# 2. we need to recursively insert that node into the left side of the tree.
# 3. if left sub tree is not present then create a new object on the left sub tree
        if data < self.key:
            if self.left:
                self.left.insert(data)
            else:
                self.left = BST(data)
# condition4: 1. if given node is greater than the root node then first check if right node is present:
# 2. we need to recursively insert that node into the right side of the tree.
# 3. if right sub tree is not present then create a new object on the right sub tree
        if data > self.key:
            if self.right:
                self.right.insert(data)
            else:
                self.right = BST(data)
#=====================================================================================================================================
#search operation
#condition1: check if tree is empty or not
# condition2: check if given node == root, print(found node)
# condition3: check if given node < root, then check again if left st is present or not(print-given node not found) then search recursively on the left side
# condition4: check if given node > root, then check again if right st is present or not(print-given node not found) then search recursively on the right side 
    def search(self, data):
        if not self.key:
            print("tree is empty")
            return
        if self.key == data:
            print("Node is found")
            return
        if data < self.key:
            if self.left:
                self.left.search(data)
            else:
                print("Given node is not present in the tree")
        if data > self.key:
            if self.right:
                self.right.search(data)
            else:
                print("Given node is not present in the tree")
#=====================================================================================================================================
# TRAVERSAL OPERATION-all of them are DFS- Depth first search
#traversal- preorder---- root, left, right
# condition1: check if the tree is empty(print-tree empty-we do not need that actually) or not(print the node-just print it)
# condition2: check if left sub tree is present then recursivly preorder on that
# condition3: check if right sub tree is present or not then recursively preorder on that
    def preorder(self):
        print(self.key, end=" ")
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
#traversal- inorder----- left-root-right 
# inorder always does accesending order
    
    def inorder(self):
        if self.left:
            self.left.inorder()
        print(self.key, end=" ")
        if self.right:
            self.right.inorder()
        
# traversal -postorder----- left-right-root
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.key, end=" ")

# only levelwise traversal is BFS-Breadth first search
# it provides the levelwise, tree level starts from 0
    
#=====================================================================================================================================
# Delete Operation
# condition1: check if the tree is empty or not: if it is empty(print-tree is empty)
# condition2: find the given node position on the tree whether it is in left side or in right side
# delete the node
    def delete(self, data, curr ):
        if not self.key:
            print("Tree is empty, cant perform deletion")
            return
        if data < self.key:
            if self.left:
                self.left= self.left.delete(data,curr)
            else: 
                print("Given node not present in the tree")
        elif data > self.key:
            if self.right:
                self.right= self.right.delete(data,curr)
            else:
                print("Given node not present in the tree ")
        else:
# node with one child 0 child or 1 child
            if self.left== None:
                temp = self.right
                if data==curr:
                    self.key = temp.key
                    self.left = temp.left
                    self.right = temp.right
                    temp = None
                    return
                    
                self= None
                return temp
            if self.right == None:
                temp = self.left
                if data==curr:
                    self.key = temp.key
                    self.left = temp.left
                    self.right = temp.right
                    temp = None
                    return
                self= None
                return temp
                
# node with 2 child
# we will have to take either node with the largest value from the left side or node with the lowest value from the right side
# here we are going to take node with lowest value i mean, right sub tree to 
# #we will use smallest node of the right sub tree and also we can take mamimum value of the left sub tree but we will work on the right sub tree so we will the minimum value of that side
            node = self.right
            while node.left:
                node = node.left
            self.key = node.key
            self.right = self.right.delete(node.key,curr)
        return self
#=====================================================================================================================================
# finding minimum value and the largest value in the tree
# we will always find the minimum value on the left side of the tree
# we will always find the maximum value on the right side of the tree
    def min(self):
        node = self
        while node.left:
            node = node.left
        print(f"Smallest value in the tree: {node.key}")

    def max(self):
        node = self
        while node.right:
            node = node.right
        print(f"Largest value in the tree: {node.key}")
#=====================================================================================================================================
# for counting how many nodes are there in a tree
t1 = BST(10)
list1 = [20,4,7,26,897,234,5]
for i in list1:
    t1.insert(i)
def count(node):
    if not node:
        return 0
    return 1+ count(node.left) + count(node.right)
# if node is the root node with no child then we can not delete the node but if it is more than the root node then we can perform the deletion operation
if count(t1)>1:
    t1.delete(10, t1.key)
else:
    print("can not perform deletion operation")
list1 = [20,4,7,26,897,234,5]
for i in list1:
    t1.insert(i)
t1.search(897)
t1.preorder()
print()
t1.inorder()
print()
t1.postorder()
t1.delete(897,10)
print()
t1.postorder()
print()
t1.min()
t1.max()
t1.postorder()
t1.delete(20,t1.key)
print()
t1.postorder()
```
